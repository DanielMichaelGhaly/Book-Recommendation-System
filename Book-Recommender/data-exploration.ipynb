```python
import kagglehub

path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax = ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
#creating a numerical representation of books which are missing

import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method = "spearman") # spearman for binary ones while pearson for continuous variables

sns.set_theme(style = "white")
plt.figure(figsize=(8,6))
heatmap = sns.heatmap(correlation_matrix, annot = True, fmt = ".2f", cmap="coolwarm", cbar_kws={"label":"Spearman correlation"})

heatmap.set_title("Correlation heatmap")

plt.show()
```

```python
books[(books["description"].isna())|
      (books["num_pages"].isna())|
      (books["average_rating"].isna())|
      (books["published_year"].isna())
]
```

```python
#Since the books with any of the 4 missing features are in total just 303 rows so like 1% of the data, its deletion will be fine

books_remaining = books[
    ~(books["description"].isna()) &
    ~(books["num_pages"].isna()) &
    ~(books["average_rating"].isna()) &
    ~(books["published_year"].isna())
]
```
